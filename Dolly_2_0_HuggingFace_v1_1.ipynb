```python
import torch
from transformers import pipeline

# use dolly-v2-12b if you're using Colab Pro+, using pythia-2.8b for Free Colab
generate_text = pipeline(model="databricks/dolly-v2-2-8b", 
                         torch_dtype=torch.bfloat16, 
                         trust_remote_code=True,
                         device_map="auto")
```

```python
generate_text("describe the what outfit I should wear today, with detailed examples of pieces of clothing, size, color. my characteristics are: black hair, 5'10, 200lb, man, brown eyes, asian. export in a json format")
```
